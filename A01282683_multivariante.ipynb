```python
import pandas as pd 
pd.set_option ("max_columns",None)
import re 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv ("antropometria-dataset.csv")
df
sistol = df['sistol']
diastol = df['diastol']
peso = df['peso']
cintura = df['cintura']
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.hist(x = sistol, bins = 10)
plt.title("Distribución de sistol")

plt.subplot(1,2,2)
plt.hist(x = diastol, bins = 10)
plt.title("Distribución de diastol")

plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.hist(x = peso, bins = 10)
plt.title("Peso")

plt.subplot(1,2,2)
plt.hist(x = cintura, bins = 10)
plt.title("Cintura")

plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.scatter(x = peso, y = cintura)
plt.xlabel("'peso'")
plt.ylabel("'cintura'")
plt.title("Diagrama de dispersión Peso-Cintura")
```
